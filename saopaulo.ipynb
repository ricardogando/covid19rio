```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
a = np.loadtxt('Dados-covid-19-estado.csv', delimiter=';', usecols=(1,2))
```

```python
plt.plot(a.T[0], label='Total')
plt.plot(a.T[1], label='Por dia')
plt.legend()
plt.xlabel('Dias')
plt.ylabel('Casos')
```

```python
derivaTotal = np.gradient(a.T[0])
derivaPordia = np.gradient(a.T[1])
```

```python
plt.plot(derivaTotal, label='Total')
plt.plot(derivaPordia, label='Por dia')
plt.legend()
for i in range(0, 49, 7):
    plt.axvline(i-1, lw=1, ls='--')
plt.xlabel('Dias')
plt.ylabel('Variao')
```
