```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
death = pd.read_csv('owid-covid-dataEUA.csv')
```

```python
vaccin = pd.read_csv('daily-covid-vaccination-doses-per-capitaUSA.csv')
```

```python
death.columns
```

```python
vaccin.columns
```

```python
usadeath = death[death.iso_code == 'USA']
```

```python
vaccin
```

```python
usavaccin = vaccin[vaccin.Code == 'USA']
bravaccin = vaccin[vaccin.Code == 'BRA']
```

```python
bradeath = death[death.iso_code == 'BRA']
```

```python
deltabra = '2021-06-28'
deltausa = '2021-'
```

```python

fig, ax = plt.subplots()

ax.plot(usadeath.date, usadeath.new_deaths_smoothed/3000, label='EUA Morte')
ax.plot(bradeath.date, bradeath.new_deaths_smoothed/3000, label='BRA')
ax.plot(usavaccin.Day, usavaccin.new_vaccinations_smoothed_per_million, '--', label='EUA Vac')
ax.plot(bravaccin.Day, bravaccin.new_vaccinations_smoothed_per_million, '--', label='BRA')
ax.legend()
plt.xlabel('Dias')
plt.ylabel('Arbitraria')

# Major ticks every 6 months.
fmt_half_year = mdates.MonthLocator(interval=6)
ax.xaxis.set_major_locator(fmt_half_year)

# Minor ticks every month.
fmt_month = mdates.MonthLocator()
ax.xaxis.set_minor_locator(fmt_month)

# Text in the x axis will be displayed in 'YYYY-mm' format.
#ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

#plt.tight_layout()
plt.savefig('mortes_e_vacinas.png', dpi=300)
```
